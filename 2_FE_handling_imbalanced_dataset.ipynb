## **Handling Imbalanced Dataset**
- Consider a binary classification problem where we need to predict yes or no. And suppose that the dataset contains 1000 datapoints - 900 yes and 100 no => which is clearly an imbalanced dataset. When we directly use this data - then we'll mostly get a biased classification as 'yes' even though it should be no.
- So we fix this problem using:
    1. **Upsampling**: extrapolate (or increase) the minority points to balance with the majority points.
    2. **Downsampling**: intrapolate (or decrease) the majority points to balance with the minority points. 

## **Create an Imbalaced dataset**

```python
# create an imbalanced dataset
import numpy as np
import pandas as pd

# set the random seed for reproducibility
np.random.seed(123)

# create a dataframe with 2 classes
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0
```

```python
n_class_0, n_class_1
```

```python
# create a dataframe using imbalanced dataset
# .normal = normal distribution
class_0 = pd.DataFrame({
    "feature_1": np.random.normal(loc = 0, scale = 1, size = n_class_0),
    "feature_2": np.random.normal(loc = 0, scale = 1, size = n_class_0),
    "target": [0]*n_class_0
})

class_1 = pd.DataFrame({
    "feature_1": np.random.normal(loc = 2, scale = 1, size = n_class_1),
    "feature_2": np.random.normal(loc = 2, scale = 1, size = n_class_1),
    "target": [1]*n_class_1
})
```

```python
# reset the indexes after concating both the dataframes
df = pd.concat([class_0, class_1]).reset_index(drop=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
df['target'].value_counts()
```

## **1. Upsampling**

- Increase the no. of 1's(100) to balance out the no. of 0's (900)
- It will try to extra-polate (increase the minority points) points to balance the minority with majority.

```python
df_minority = df[df['target'] == 1]
df_majority = df[df['target'] == 0]
```

```python
from sklearn.utils import resample

df_minority_upsampled = resample(df_minority, 
                                 replace=True, # sample with replacemet
                                 n_samples = len(df_majority), 
                                 random_state=42
                                 )
```

```python
df_minority_upsampled.shape
```

```python
df_minority_upsampled.head()
```

```python
df_upsampled = pd.concat([df_majority, df_minority_upsampled])
```

```python
df_upsampled
```

```python
df_upsampled['target'].value_counts()
```
