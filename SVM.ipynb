## **Support Vector Machine (SVM)**

This ML algorithm is used for both classification and regression problems

1. SVC (Support Vector Classifier)
2. SVR (Support Vector Regressor)

**1. SVC (Support Vector Classifier)**



- We try to create a best fit line like logistic regression along with that we try to create 2 marginal planes (1 above and 1 below the best fit line) such that they are equidistant from the best fit line and also we try to maximise the distance from the best fit line to the nearest points (to the best fit line) to classify correctly



- Also whichever d>d' we choose that "d" for the making the classification

- The nearest point to the marginal lines are called as **Support Vectors**



- For 3 dimensional problems we create a best fit plane along with 2 marginal planes and here also we try to maximise the distance of the marginal planes from the best fit plane

**Soft Margin**



- In real world scenarios, the points x1 and x2 will be overlapped and hence we dont get a clear cut best fit line, and also it will be prone to minor errors, this is called as soft margin

**Hard Margin**



- If we a scenario where we are clearly able to distinguish the x1 and x2 for creating a best fit line, then its called as hard margin

**Support Vector Machine - Maths Intution**




d = -ve below the plane

d = +ve above the plane




we have the make sure that the distance between the marginal plane is maximised











### **Support Vector Classifier Implementation**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
```

```python
# create a synthetic dataset
X, y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=1, n_redundant=0)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)
```

```python
pd.DataFrame(X)[0]
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y)
```

```python
# better way
df = pd.DataFrame(X, columns=['feature_1', 'feature_2'])
df['target']=y
```

```python
sns.scatterplot(x='feature_1', y='feature_2', hue='target', data=df)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='linear')
```

```python
svc.fit(X_train, y_train)
```

```python
# prediction

y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print("Accuracy: ")
print(accuracy_score(y_test, y_pred))

print("Classification report: ")
print(classification_report(y_test, y_pred))

print("Confusion matrix: ")
print(confusion_matrix(y_test, y_pred))
```

```python
svc = SVC(kernel='poly')
```

```python
svc.fit(X_train, y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
print("Accuracy score: ", accuracy_score(y_test, y_pred))
```

```python
svc = SVC(kernel='sigmoid')
```

```python
svc.fit(X_train, y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
print("Accuracy score: ", accuracy_score(y_test, y_pred))
```

```python
svc = SVC(kernel='rbf')
```

```python
svc.fit(X_train, y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
print("Accuracy score: ",accuracy_score(y_test, y_pred))
```

Hence out of all the types of kernel, rbf is the best here which gave an accuracy of 98%

Hence use the appropriate kernels for appropriate patterns in the dataset

**Hyper Parameter Tuning**

```python
from sklearn.model_selection import GridSearchCV

# defining parameter range
param_grid = {
                'C': [0.1, 1, 10, 100, 1000],
                'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
                'kernel': ['rbf']
                }
```

```python
grid = GridSearchCV(SVC(), param_grid=param_grid, refit=True, cv=5, verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
# predictions
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
