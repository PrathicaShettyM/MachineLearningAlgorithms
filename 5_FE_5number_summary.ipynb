```python
import numpy as np
```

```python
lst_marks = [45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74]
minimum, Q1, median, Q3, maximum = np.quantile(lst_marks, [0, 0.25, 0.50, 0.75, 1.0])
```

```python
minimum, Q1, median, Q3, maximum
```

#### **Calculating Outliers**

```python
IQR = Q3 - Q1
print(IQR)
```

```python
lower_fence = Q1 - 1.5 * IQR
higher_fence = Q3 + 1.5 * IQR 
```

```python
lower_fence
```

```python
higher_fence
```

So hence anything lower than lower_fence(1.5) and anything higher than higher_fence(141.5) is considered as an outlier

```python
import seaborn as sns
```

```python
sns.boxplot(lst_marks)
```

There is no outlier in this dataset, so we are not getting any dots in the box plot
