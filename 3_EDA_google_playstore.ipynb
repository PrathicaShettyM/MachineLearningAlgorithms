```python
df.isnull().sum() 
```

**1. Handle Reviews**

Convert it to a numeric type to feed it to the model

```python
df['Reviews'].unique()
```

```python
df['Reviews'].str.isnumeric().sum()
```

**Observation**

Hence we see that almost all the values (except 1 record) are numeric string types but the one record which is 30M which is definitely not numeric

```python
# find that non-numeric value
df[~df['Reviews'].str.isnumeric()] 
```

```python
# make a copy of the dataframe
df_copy = df.copy()
```

```python
df_copy = df_copy.drop(df_copy.index[10472])
```

```python
# check if the record is dropped
df_copy[~df_copy['Reviews'].str.isnumeric()]

# It will be empty
```

```python
df_copy['Reviews'] = df_copy['Reviews'].astype(int)
```

**2. Handle the Size**

Remove those 'M' or 'k' and convert them to proper integer formats

```python
df_copy['Size'].unique()
```

**Observation**

Since most of the records are in 'k' so its better to convert those 'M' to 'k'. for example if any value is like 19M, then its value in numerical format is 19,000,000 then if we remove 'k' it becomes 19000k

```python
df_copy['Size'] = df_copy['Size'].str.replace('M', '000')
df_copy['Size'] = df_copy['Size'].str.replace('k', '') # keep 'k' as the default value so it shd be treated as blank here
df_copy['Size'] = df_copy['Size'].replace('Varies with device', np.nan)

# next convert them to float types
df_copy['Size'] = df_copy['Size'].astype(float)
```

```python
df_copy.info()
```

**3. Handle Installs and Price**

we have to remove '+', '$' and strings like "Free" and "Everyone"

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

**Observation**

we have to remove '+', '$' and strings like "Free" and "Everyone"

```python
chars_to_remove = ['+', ',', '$'] # special characters to remove

cols_to_clean = ['Installs', 'Price']

for item in chars_to_remove:
    for cols in cols_to_clean:
        df_copy[cols] = df_copy[cols].str.replace(item,'')
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python
df_copy['Installs'] = df_copy['Installs'].astype(int)
df_copy['Price'] = df_copy['Price'].astype(float)
```

```python
df_copy.info()
```

**4. Handle Last Updated**

Handle Date types

```python
df.head(2)
```

```python
df_copy['Last Updated'] = pd.to_datetime(df_copy['Last Updated'])

df_copy['Day'] = df_copy['Last Updated'].dt.day
df_copy['Month'] = df_copy['Last Updated'].dt.month
df_copy['Year'] = df_copy['Last Updated'].dt.year
```

```python
df_copy.info()
```

```python
df_copy.to_csv('google_cleaned_1.csv')
```

### **Explore the Data**

```python
numerical_features = [feature for feature in df_copy.columns if df_copy[feature].dtype != 'O']
categorical_features = [feature for feature in df_copy.columns if df_copy[feature].dtype == 'O']

# print columns
print('We have {} numerical features: {}'.format(len(numerical_features), numerical_features))
print('\nWe have {} categorical features: {}'.format(len(categorical_features), categorical_features))
```

#### **Features information**

0.   App            : name of the app
1.   Category       : category of the app
2.   Rating         : ratings on playstore
3.   Reviews        : reviews on the app
4.   Size           : size of the app
5.   Installs       : no. of installs of the app
6.   Type           : is the app free/paid
7.   Price          : price if exists (0 if its free)
8.   Content Rating : appropriate target audience of the app
9.   Genres         : genre of the app
10.  Last Updated   : last updated date
11.  Current Ver    : current verison of the app
12.  Android Ver    : minimum android version to run the app

```python
# proportion of count data on categorical columns

for col in categorical_features:
    print(df[col].value_counts(normalize=True)*100) # this gives us values in the form of %
    print('--------------------------------------------------------')
```

**Visualisation Diagrams**

```python
# Proportion of count data on numerical columns

plt.figure(figsize=(15,15))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1)

for i in range(0, len(numerical_features)):
    plt.subplot(5, 3, i+1)
    sns.kdeplot(x = df_copy[numerical_features[i]], shade=True, color='r') # kernel density estimation plot
    plt.xlabel(numerical_features[i])
    plt.tight_layout()
```

**Observation**

- **Rating** and **Year** are `left skewed`
- While **Reviews**, **Size**, **Installs** and **Price** are `right skewed`

Hence we can say that the outliers are present

```python
# Proportion of count data on categorical features

plt.figure(figsize=(20, 15))
plt.suptitle('Univariate Analysis of Categorical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1)
category = ['Type', 'Content Rating']

for i in range(0, len(category)):
    plt.subplot(2, 2, i+1)
    sns.countplot(x = df[category[i]], palette="Set2") # count plot is used with categories
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
```
